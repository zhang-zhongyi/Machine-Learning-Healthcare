```python
#import modules
import pandas as pd
from sklearn.linear_model import LinearRegression 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score as roc
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV as rcv
import sklearn.preprocessing as pp
import numpy as np
```

```python
#importing and reading the csv
credit_train = pd.read_csv("~/Box/teaching/2019_fall_msbi_33200/datasets/credit_card/training_credit_data.csv")
credit_test = pd.read_csv("~/Box/teaching/2019_fall_msbi_33200/datasets/credit_card/test_credit_data.csv")

print(credit_train)
```

```python

le = pp.LabelEncoder()
credit_train.Gender= le.fit_transform(credit_train.Gender)
credit_train.Student = le.fit_transform(credit_train.Student)
credit_train.Married = le.fit_transform(credit_train.Married)
credit_train.Ethnicity = le.fit_transform(credit_train.Ethnicity)
print(credit_train)

credit_test.Gender= le.fit_transform(credit_test.Gender)
credit_test.Student = le.fit_transform(credit_test.Student)
credit_test.Married = le.fit_transform(credit_test.Married)
credit_test.Ethnicity = le.fit_transform(credit_test.Ethnicity)
```

```python

X_train = credit_train.drop(columns = ["Balance"])
X_test = credit_test.drop(columns = ["Balance"])
y_train = credit_train.Balance
y_test = credit_test.Balance


```

```python
# Linear Regression Models
lm_model = LinearRegression()
lm_model.fit(X_train, y_train)

print(lm_model)
```

```python

test_predictions = lm_model.predict(X_test)
R = np.corrcoef(test_predictions, y_test)
R2 = R**2
print(R2)
```
